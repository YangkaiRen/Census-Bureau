```python
test_X = test_df.drop('income', axis=1)
```

```python
df = pd.concat([train_df, test_X], ignore_index=True)
```

```python
df[df == '?'] = np.nan
```

```python
df.info()
```

The dataset contained null values, both numerical and categorical values. The categorical values were both nominal and ordinal. The data had redundant columns as well.


```python
#The dependent column ‘income’ which is to be predicted has been replaced with 0 and 1 and hence convert the problem to a dichotomous classification problem.

df['income'].replace({'<=50K':0,'>50K':1},inplace=True)
```

```python
df.info()
```

```python
#To check the correlation between a binary variable and continuous variables, the point biserial correlation has been used. 

import scipy.stats as stats
nums = df.select_dtypes(include=['int'])

# nums = ['age','capital.loss','capital.gain','hours.per.week','fnlwgt']
for i in nums.columns:

    print(i,':', stats.pointbiserialr(df[df['income'].notnull()]['income'], df[df['income'].notnull()][i])[0])
```

```python
# fnlwgt showed negative correlation. So droped fnlwgt.
#education.num which is an ordinal representation of ‘education’, remove it too.

df = df.drop('fnlwgt',axis=1)
df = df.drop('education.num', axis=1)
```

---------------------------------Numerical Columns-----------------------------------

Handling Numerical Columns. I select the numerical columns using the select_dtypes function.

```python
num_attributes = df.select_dtypes(include=['int'])
num_columns = list(num_attributes.columns)
num_columns
```

```python
plt.figure(figsize = (10, 10))
for i in range(4):
    plt.subplot(2,2,i+1)
    sns.distplot(df[df['income'].notnull()][num_columns[i]], color = 'blue', hist=False)
    sns.distplot(df[df['income'].isnull()][num_columns[i]], color = 'yellow', hist=False)
plt.tight_layout()
```

The values are on different scales. Many machine learning models require the values to be on the same scale. We will use StandardScaler from the sklearn library to scale the features.

---------------------------------Categorical Columns--------------------------------

```python
cat_attributes = df.select_dtypes(include=['object'])
cat_columns = list(cat_attributes.columns)
cat_columns
```

```python
plt.figure(figsize = (18,30))
for i,j in enumerate(cat_columns):
    plt.subplot(4,2, i+1)
    sns.countplot(x=cat_attributes[df['income'].notnull()][j], hue=df[df['income'].notnull()]['income'])
    plt.xticks(rotation=90)
plt.tight_layout()
```
